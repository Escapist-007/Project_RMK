### Creating a new column "Micro_time_int" as int type data 

```python
df.dtypes
```

```python
df['Micro_time_int'] = df.Micro_time.str.replace(':','.').astype(float)
```

### Applying Aggregate functions

```python
df.groupby('Day').Micro_time_int.agg(['count','max','min','mean','median', 'std',lambda x:x.value_counts().index[0]])
```

```python
df.groupby('Day').Micro_time_int.mean().plot(kind='bar')
```

```python
df.groupby('Day').Micro_time_int.agg(['max','min','mean','median',lambda x:x.value_counts().index[0]]).plot(kind='bar', yticks = np.arange(0,9,0.5),figsize=(15,4))
```
